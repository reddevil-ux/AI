```python
!pip install -q python-telegram-bot transformers torch Pillow nest_asyncio > /dev/null 2>&1
import nest_asyncio
nest_asyncio.apply()
import logging
import torch
import random
import re
import asyncio
from typing import List, Dict, Set
from PIL import Image
from io import BytesIO
from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes, CallbackQueryHandler
from transformers import BlipProcessor, BlipForConditionalGeneration, GPT2LMHeadModel, GPT2Tokenizer

a = "8484962267:AAFfAoLvBWdzWd8UJtW8aW0BLNoH0LmDHf8"
logging.basicConfig(level=logging.WARNING)
b = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
c = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
d = GPT2Tokenizer.from_pretrained("distilgpt2")
e = GPT2LMHeadModel.from_pretrained("distilgpt2")
f = torch.device("cuda" if torch.cuda.is_available() else "cpu")
c = c.to(f)
e = e.to(f)
d.pad_token = d.eos_token
g = {}
h = 2
i = 2
j = 3
k = 20

l = [
"A young girl laughing while holding a colorful balloon in a park.",
"A man wearing a red hoodie walking his dog on a city sidewalk.",
"A group of friends sitting on a bench eating ice cream in summer.",
"A woman reading a book under a big oak tree in a garden.",
"A little boy playing with a toy car on a green grass lawn.",
"A grandmother hugging her granddaughter at a family dinner.",
"A chef wearing a white hat cooking pasta in a restaurant kitchen.",
"A student writing notes in a notebook at a library table.",
"A couple holding hands walking along a sandy beach at sunset.",
"A musician playing a guitar on a street corner for passersby.",
"A firefighter wearing a helmet carrying a small child to safety.",
"A teacher explaining a lesson to a class of young children.",
"A photographer taking pictures of a wedding with a professional camera.",
"A hiker with a backpack climbing a steep mountain trail.",
"A barista making a cup of coffee with milk foam in a café.",
"A dancer performing a spin on a wooden stage with bright lights.",
"A doctor wearing a white coat checking a patient's heartbeat.",
"A painter holding a brush painting a landscape on a canvas.",
"A child blowing out candles on a birthday cake with friends.",
"A cyclist wearing a helmet riding a bike on a country road.",
"A mother pushing a baby stroller through a busy shopping mall.",
"A father teaching his son how to ride a bicycle in a backyard.",
"A group of teenagers playing basketball on an outdoor court.",
"A nurse giving a vaccine to a young child in a clinic.",
"A pilot wearing a uniform standing next to a small airplane.",
"A artist drawing a portrait of a woman with a pencil on paper.",
"A waiter carrying a tray of drinks to a table in a restaurant.",
"A swimmer diving into a blue swimming pool at a sports center.",
"A skier sliding down a snowy mountain slope with ski poles.",
"A baker putting fresh bread into a hot oven in a bakery.",
"A journalist interviewing a politician with a microphone in hand.",
"A yoga instructor leading a class on a mat in a studio.",
"A construction worker wearing a vest lifting a metal beam.",
"A hairdresser cutting a woman's hair with scissors in a salon.",
"A fisherman sitting on a wooden pier fishing with a rod in a lake.",
"A actress reciting lines on a theater stage with a script.",
"A tailor sewing a dress with a needle and thread on a sewing machine.",
"A runner jogging through a forest trail with leaves on the ground.",
"A makeup artist applying lipstick to a model's lips backstage.",
"A gardener planting flowers in a flower bed with a small trowel.",
"A police officer directing traffic at a busy city intersection.",
"A singer performing a song into a microphone at a music concert.",
"A scientist looking through a microscope in a laboratory.",
"A babysitter playing with a group of toddlers in a playroom.",
"A delivery driver handing a package to a customer at a door.",
"A sculptor carving a statue from a block of marble with a chisel.",
"A tennis player hitting a ball with a racket on a tennis court.",
"A writer typing on a laptop at a desk near a window with sunlight.",
"A lifeguard sitting on a high chair watching swimmers in a pool.",
"A florist arranging a bouquet of roses and lilies in a vase.",
"A mechanic fixing a car engine with a wrench in a garage.",
"A poet reading a poem to a small audience in a bookstore.",
"A skateboarder doing a trick on a ramp in a skate park.",
"A judge wearing a black robe sitting in a courtroom with a gavel.",
"A farmer harvesting wheat with a combine in a golden field.",
"A DJ mixing music on a turntable at a night club with neon lights.",
"A astronomer looking at the stars through a telescope on a roof.",
"A seamstress hemming a pair of pants with a sewing needle.",
"A rock climber climbing a vertical rock wall with ropes and harness.",
"A pastry chef decorating a cake with white frosting and berries.",
"A translator speaking into a headset at an international conference.",
"A golfer swinging a club at a golf ball on a green course.",
"A potter shaping clay into a bowl on a pottery wheel in a studio.",
"A flight attendant serving drinks to passengers on an airplane.",
"A personal trainer helping a client lift weights in a gym.",
"A calligrapher writing words with a brush on rice paper.",
"A sailor steering a boat with a wheel on the open ocean.",
"A butcher cutting meat with a knife on a wooden cutting board.",
"A violinist playing a melody with a bow on a violin at a concert.",
"A interior designer arranging furniture in a living room with a sofa.",
"A miner wearing a headlamp digging in a dark underground tunnel.",
"A pastry cook making croissants with dough in a bakery kitchen.",
"A stand-up comedian telling jokes to a laughing audience in a club.",
"A geologist examining a rock sample with a magnifying glass.",
"A dressmaker fitting a wedding dress on a bride in a boutique.",
"A roofer laying shingles on a house roof with a hammer.",
"A cellist playing a slow song on a cello in a music hall.",
"A landscape designer planning a garden with plants and a fountain.",
"A blacksmith forging a horseshoe with a hammer and anvil.",
"A comedian performing a skit with a partner on a comedy stage.",
"A meteorologist pointing at a weather map on a news broadcast.",
"A shoemaker repairing a shoe with glue and a hammer in a shop.",
"A flutist playing a tune on a flute in a small music room.",
"A event planner setting up chairs for a wedding reception in a hall.",
"A welder welding metal pipes with a welding torch in a workshop.",
"A magician performing a card trick for a group of amazed children.",
"A cartographer drawing a map with a pen on a large piece of paper.",
"A glassblower shaping glass into a vase with a blowpipe in a studio.",
"A ventriloquist performing with a puppet on a stage for an audience.",
"A biologist studying a plant sample in a petri dish in a lab.",
"A locksmith opening a door with a set of keys in a hand.",
"A small black cat sleeping on a soft red sofa near a window.",
"A golden retriever dog playing with a ball in a green grass field.",
"A colorful parrot flying above a tropical forest with green leaves.",
"A white rabbit eating a carrot in a garden with fresh vegetables.",
"A brown squirrel climbing a tall oak tree with a nut in its mouth.",
"A playful puppy chasing its tail on a wooden floor in a house.",
"A majestic eagle flying high in a blue sky with white clouds.",
"A cute kitten playing with a ball of yarn on a living room carpet.",
"A loyal dog waiting for its owner at a white picket fence.",
"A bright red cardinal bird sitting on a tree branch in winter.",
"A fluffy white sheep grazing on green grass in a mountain meadow.",
"A fast cheetah running across a dry savanna in Africa.",
"A curious monkey swinging from tree to tree in a jungle.",
"A graceful swan swimming on a calm blue lake with lily pads.",
"A tiny hamster running on a wheel in a glass cage with bedding.",
"A powerful lion resting in the shade of a tree on a hot savanna.",
"A beautiful peacock spreading its colorful tail feathers in a zoo.",
"A friendly dolphin jumping out of the blue ocean water near a boat.",
"A slow tortoise walking on a sandy beach near the ocean waves.",
"A sleek black panther hiding in the tall grass of a rainforest.",
"A small goldfish swimming in a glass bowl with colorful pebbles.",
"A strong horse galloping across a green field with a rider.",
"A cute penguin sliding on ice in a snowy Antarctic landscape.",
"A wise old owl sitting on a branch at night with big eyes.",
"A playful otter floating on its back in a river with a stone.",
"A large elephant drinking water from a river with its long trunk.",
"A tiny ladybug crawling on a red rose petal in a garden.",
"A fierce tiger walking through tall grass in a Asian jungle.",
"A gentle giraffe eating leaves from a tall acacia tree in Africa.",
"A small hedgehog rolling into a ball on a forest floor with leaves.",
"A colorful butterfly flying from flower to flower in a meadow.",
"A loyal wolf howling at the full moon in a snowy forest.",
"A cute guinea pig eating lettuce in a cage with hay and toys.",
"A fast falcon diving down to catch a small bird in the sky.",
"A fluffy cat stretching on a sunny windowsill with a view.",
"A big brown bear fishing for salmon in a fast-moving river.",
"A tiny ant carrying a crumb of bread on a kitchen counter.",
"A beautiful flamingo standing on one leg in a pink salt lake.",
"A playful seal swimming in a cold ocean near a rocky shore.",
"A large hippopotamus bathing in a muddy river in Africa.",
"A small sparrow eating seeds from a bird feeder in a backyard.",
"A sleek grey wolf walking through a forest with pine trees.",
"A cute chihuahua dog sitting on a couch with a blanket.",
"A powerful hawk flying above a mountain range with cliffs.",
"A tiny snail crawling on a green leaf with a spiral shell.",
"A majestic peacock walking on a grassy lawn in a park.",
"A friendly golden retriever swimming in a lake with a stick.",
"A small red fox hunting for rabbits in a snowy field.",
"A graceful deer standing in a forest with fall leaves on the ground.",
"A tiny frog jumping on a lily pad in a pond with water lilies.",
"A large gorilla sitting on a rock in a jungle with its baby.",
"A colorful fish swimming in a coral reef with tropical plants.",
"A cute bunny hopping through a field of yellow daisies in spring.",
"A fierce shark swimming in the deep blue ocean with fish around.",
"A gentle cow eating grass in a green meadow with a bell.",
"A tiny hummingbird drinking nectar from a red flower with its beak.",
"A large alligator lying in the sun on a muddy riverbank.",
"A fluffy rabbit sitting in a basket with straw and a carrot.",
"A fast zebra running across a savanna with black and white stripes.",
"A cute ferret playing with a toy in a living room with a rug.",
"A powerful rhino walking through a grassy plain in Africa.",
"A beautiful blue jay bird flying with a twig in its beak.",
"A friendly parakeet sitting on its owner's finger with a seed.",
"A slow sloth hanging from a tree branch in a rainforest.",
"A large moose walking through a forest with antlers on its head.",
"A tiny firefly glowing in the dark near a bush with leaves.",
"A fierce coyote howling in the desert with sand dunes.",
"A gentle sheepdog herding sheep in a green field with a farmer.",
"A cute duckling following its mother across a pond with water.",
"A large whale swimming in the ocean with a spray of water from its blowhole.",
"A tiny gecko climbing on a wall with small feet.",
"A friendly octopus swimming in a tank with coral and rocks.",
"A fast rabbit running away from a fox in a green field.",
"A cute hamster eating a sunflower seed in a small cage.",
"A powerful eagle catching a fish from a lake with its talons.",
"A beautiful butterfly resting on a purple flower in a garden.",
"A loyal dog barking at a squirrel in a tree in a backyard.",
"A tiny ladybug flying from a leaf to a flower in a meadow.",
"A fierce lion roaring at other animals on a African savanna.",
"A gentle giraffe bending down to drink water from a pond.",
"A playful puppy chewing on a bone in a dog bed with a toy.",
"A sleek black cat walking on a fence at night with the moon.",
"A colorful parrot talking to its owner in a cage with toys.",
"A cute penguin waddling on ice with other penguins in a group.",
"A wise old owl hunting for mice in a forest at night.",
"A friendly dolphin playing with a ball in a marine park pool.",
"A small goldfish blowing bubbles in a glass bowl with water.",
"A cup of hot tea with lemon on a wooden table near a window.",
"A black laptop open on a desk with a plant and a notebook.",
"A red bicycle parked next to a brick wall with green ivy.",
"A white mug with a blue pattern filled with coffee and milk.",
"A old book with a leather cover lying on a wooden shelf.",
"A silver watch on a black leather band on a glass table.",
"A colorful painting hanging on a white wall in a living room.",
"A wooden chair with a cushion next to a fireplace with a fire.",
"A glass vase with red roses and green leaves on a dining table.",
"A blue backpack with a water bottle and a notebook inside.",
"A stone path leading through a garden with colorful flowers.",
"A wooden bridge crossing a small stream with clear water.",
"A snow-covered mountain peak with a blue sky in the background.",
"A sandy beach with white waves crashing on the shore at sunrise.",
"A green forest with tall trees and sunlight through the leaves.",
"A small cottage with a red roof in a field of yellow sunflowers.",
"A city street with tall buildings and cars driving on the road.",
"A quiet lake with calm water and trees reflecting on the surface.",
"A rocky coastline with waves hitting the rocks and seagulls flying.",
"A grassy meadow with wildflowers and a small stream running through.",
"A metal lantern with a candle inside hanging from a tree branch.",
"A porcelain plate with a floral design on a white kitchen counter.",
"A black pen with a silver clip lying on a white piece of paper.",
"A woven basket with fresh fruits and vegetables on a wooden floor.",
"A glass jar filled with colorful candies on a coffee table.",
"A wooden cutting board with a knife and fresh vegetables on it.",
"A silk scarf with a floral pattern folded on a velvet cushion.",
"A brass key hanging on a wooden key rack near a front door.",
"A ceramic bowl with fruit on a glass coffee table in a living room.",
"A woolen blanket with a checkered pattern on a brown leather sofa.",
"A mountain trail with rocks and trees leading to a scenic viewpoint.",
"A desert landscape with sand dunes and a clear blue sky at sunset.",
"A tropical beach with palm trees and turquoise water near the shore.",
"A forest path with fallen leaves and mushrooms growing on the ground.",
"A village street with small houses and a church with a steeple.",
"A river winding through a green valley with mountains in the distance.",
"A meadow with tall grass and butterflies flying in the summer sun.",
"A cliff with a view of the ocean and waves crashing below.",
"A garden with a fountain and colorful flowers around it.",
"A snowy forest with pine trees and animal tracks in the snow.",
"A plastic water bottle with a blue cap on a green park bench.",
"A paper notebook with a spiral binding and a pen on top of it.",
"A wooden spoon with flour on it lying on a baking tray in a kitchen.",
"A glass wine glass with red wine on a white tablecloth at a dinner.",
"A metal fork and knife on a ceramic plate with a napkin folded beside.",
"A cotton towel with a striped pattern hanging on a bathroom rack.",
"A plastic toy car with red wheels on a carpet in a child's room.",
"A wooden puzzle with colorful pieces spread on a coffee table.",
"A silk pillow with a embroidery design on a white bed with sheets.",
"A brass candlestick with a white candle burning on a wooden shelf.",
"A glass fish tank with colorful fish and green plants inside it.",
"A metal watering can with water on a garden path near flowers.",
"A plastic bucket and shovel on a sandy beach near the water.",
"A wooden bird feeder with seeds hanging from a tree in a backyard.",
"A glass picture frame with a family photo on a bedroom dresser.",
"A metal flashlight with a black handle on a camping table in the woods.",
"A paper bag with groceries on a kitchen counter near a sink.",
"A woolen hat with a pom-pom on top lying on a winter coat.",
"A leather wallet with a few cards and cash inside on a table.",
"A plastic phone case with a floral design on a smartphone on a desk.",
"A wooden guitar leaning against a wall in a music room with a chair.",
"A metal telescope on a tripod looking at the stars on a roof at night.",
"A ceramic mug with a funny quote on it filled with hot chocolate.",
"A woven rug with a geometric pattern on a wooden floor in a living room.",
"A glass jar with honey and a wooden dipper on a breakfast table.",
"A metal baking pan with cookies baking in a hot oven in a kitchen.",
"A paper kite with a long tail flying in a blue sky with wind.",
"A plastic frisbee lying on a green grass lawn in a park.",
"A wooden bench with a metal frame under a big tree in a garden.",
"A glass shower door with water droplets on it in a bathroom.",
"A metal doorknob with a brass finish on a wooden front door.",
"A ceramic flower pot with a green plant on a windowsill with sunlight.",
"A plastic straw in a glass of lemonade on a patio table in summer.",
"A wooden ladder leaning against a wall in a garage with tools.",
"A metal trash can with a plastic bag inside near a kitchen sink.",
"A paper calendar with marked dates on a wall near a desk.",
"A woolen sweater with a cable knit pattern folded on a chair.",
"A leather belt with a metal buckle on a bedroom bed with a blanket.",
"A plastic raincoat hanging on a hook near a front door on a rainy day.",
"A wooden cutting board with a loaf of bread and a knife on it.",
"A glass jar with pickles and a metal lid on a kitchen shelf.",
"A metal spoon in a bowl of soup on a dinner plate with a napkin.",
"A paper napkin with a printed design on a dining table at a restaurant.",
"A plastic toy bear with a red bow on its neck on a child's bed.",
"A wooden bookend with a stone design holding books on a shelf.",
"A glass candle holder with a scented candle burning on a coffee table.",
"A metal bike lock attached to a bicycle parked on a city street.",
"A paper map folded on a table with a pen marking a route on it."
]

def m(n,o):
    p=re.compile(r'[^\w\s]')
    q=p.sub('',n.lower()).split()
    r=p.sub('',o.lower()).split()
    s=set(q)
    t=set(r)
    u=s.intersection(t)
    return len(u)

def v(w):
    x=re.sub(r'[^\w\s]','',w.lower()).split()
    if len(x)<j or len(x)>k:
        return False
    y=1
    for z in range(1,len(x)):
        if x[z]==x[z-1]:
            y+=1
            if y>i:
                return False
        else:
            y=1
    for aa in x:
        if len(aa)<2 and aa not in ['a','an','in','on','at','to']:
            return False
    return True

async def ab(ac,ad):
    ae=[[InlineKeyboardButton("Угадать описание",callback_data="mode1")],[InlineKeyboardButton("Написать свое описание",callback_data="mode2")]]
    af=InlineKeyboardMarkup(ae)
    await ac.message.reply_text("Выбери игровой режим:",reply_markup=af)

def ag(ah):
    ai=Image.open(BytesIO(ah))
    if ai.mode!="RGB":
        ai=ai.convert("RGB")
    ai.thumbnail((400,400))
    aj=b(images=ai,return_tensors="pt").to(f)
    ak=5
    for _ in range(ak):
        with torch.no_grad():
            al=c.generate(**aj,max_length=30,num_beams=3,repetition_penalty=1.5,early_stopping=True)
        am=b.decode(al[0],skip_special_tokens=True)
        if am:
            an=am.split()
            an[0]=an[0].capitalize()
            am=' '.join(an)
        if not am.endswith('.'):
            am+='.'
        if v(am):
            return am
    return "A scene with various elements in a natural setting."

def ao(ap,aq=4):
    ar=[]
    as_=len(ap.split())
    at=[]
    for au in l:
        if (au.lower()!=ap.lower() and m(ap,au)<=h and abs(len(au.split())-as_)<=1 and v(au)):
            at.append(au)
    random.shuffle(at)
    if len(at)>=aq:
        ar=at[:aq]
    else:
        ar=at.copy()
        av=[aw for aw in l if aw not in ar and aw.lower()!=ap.lower() and m(ap,aw)<=h and v(aw)]
        random.shuffle(av)
        ar.extend(av[:aq-len(ar)])
    return ar[:aq]

def ax(ay,az,ba,bb=4):
    bc=ao(ay,bb)
    bd=[ay]+bc
    be=random.randint(0,len(bd)-1)
    if bd[0]==ay:
        bd[0],bd[be]=bd[be],bd[0]
    else:
        for bf,bg in enumerate(bd):
            if bg==ay:
                bd[bf],bd[be]=bd[be],bd[bf]
                break
    if az not in g:
        g[az]={}
    g[az][ba]={"bh":ay,"bi":bd}
    bj=[]
    bk=[]
    for bl in range(len(bd)):
        bk.append(InlineKeyboardButton(f"{bl+1}",callback_data=f"ans_{az}_{ba}_{bl}"))
        if (bl+1)%5==0:
            bj.append(bk)
            bk=[]
    if bk:
        bj.append(bk)
    bm=InlineKeyboardMarkup(bj)
    bn="\n".join([f"{bo+1}. {bp}" for bo,bp in enumerate(bd)])
    return bm,bd,bn

async def bq(br,bs):
    bt=br.callback_query
    await bt.answer()
    bu=bt.data
    bv=br.effective_user.id
    if bu.startswith("mode"):
        g[bv]={"bw":bu}
        if bu=="mode1":
            bx=[]
            by=[]
            for bz in range(1,21):
                by.append(InlineKeyboardButton(str(bz),callback_data=f"fake{bz}"))
                if bz%5==0:
                    bx.append(by)
                    by=[]
            ca=InlineKeyboardMarkup(bx)
            await bt.edit_message_text("Выбери количество неверных вариантов (1-20):",reply_markup=ca)
        else:
            await bt.edit_message_text("Режим активирован. Отправь фото, затем напиши свое описание на английском.")
    elif bu.startswith("fake"):
        cb=int(bu[4:])
        g[bv]["cc"]=cb
        await bt.edit_message_text(f"Режим активирован. {cb} неверных вариантов. Отправь фото для начала игры.")
    elif bu.startswith("ans"):
        cd=bu.split('_')
        if len(cd)!=4:
            return
        ce=int(cd[1])
        cf=int(cd[2])
        cg=int(cd[3])
        if ce not in g or cf not in g[ce]:
            await bt.edit_message_text("Сессия истекла. Выбери режим заново.")
            return
        ch=g[ce][cf]
        ci=ch["bh"]
        cj=ch["bi"]
        ck=cj[cg]
        cl=(ck==ci)
        cm=[]
        for cn,co in enumerate(cj):
            if co==ci:
                cm.append(f"✅ {cn+1}. {co}")
            elif cn==cg and not cl:
                cm.append(f"❌ {cn+1}. {co}")
            else:
                cm.append(f"- {cn+1}. {co}")
        cp="\n".join(cm)
        cq=f"Правильно!\n\n{cp}" if cl else f"Неправильно!\n\n{cp}"
        await bt.edit_message_text(cq)
        del g[ce][cf]

async def cr(cs,ct):
    cu=cs.effective_user.id
    if cu not in g or "bw" not in g[cu]:
        cv=[[InlineKeyboardButton("Угадать описание",callback_data="mode1")],[InlineKeyboardButton("Написать свое описание",callback_data="mode2")]]
        cw=InlineKeyboardMarkup(cv)
        await cs.message.reply_text("Сначала выбери игровой режим.",reply_markup=cw)
        return
    cx=g[cu]["bw"]
    cy=cs.message.photo
    if len(cy)>1:
        cz=await cy[-2].get_file()
    else:
        cz=await cy[-1].get_file()
    da=await cz.download_as_bytearray()
    db=ag(da)
    if cx=="mode1":
        if "cc" not in g[cu]:
            dc=[]
            dd=[]
            for de in range(1,21):
                dd.append(InlineKeyboardButton(str(de),callback_data=f"fake{de}"))
                if de%5==0:
                    dc.append(dd)
                    dd=[]
            df=InlineKeyboardMarkup(dc)
            await cs.message.reply_text("Сначала выбери количество неверных вариантов (1-20):",reply_markup=df)
            return
        dg=g[cu]["cc"]
        dh=await cs.message.reply_text("Обрабатываю фото...")
        di=dh.message_id
        dj,dk,dl=ax(db,cu,di,dg)
        dm=f"Выбери правильное описание:\n\n{dl}"
        await dh.edit_text(dm,reply_markup=dj)
    elif cx=="mode2":
        g[cu]["dn"]=db
        await cs.message.reply_text("Фото обработано. Напиши свое описание на английском.")

async def do(dp,dq):
    dr=dp.effective_user.id
    ds=dp.message.text
    if dr not in g or g[dr].get("bw")!="mode2" or "dn" not in g[dr]:
        dt=[[InlineKeyboardButton("Угадать описание",callback_data="mode1")],[InlineKeyboardButton("Написать свое описание",callback_data="mode2")]]
        du=InlineKeyboardMarkup(dt)
        await dp.message.reply_text("Сначала активируй режим Написать свое описание.",reply_markup=du)
        return
    dv=g[dr]["dn"]
    dw=ds.strip().capitalize()
    if not dw.endswith('.'):
        dw+='.'
    dx=m(dv,dw)
    dy=len(dv.split())
    dz=len(dw.split())
    ea=abs(dz-dy)<=5
    if dw.lower()==dv.lower() or (dx>=int(dy*0.7) and ea):
        eb=f"Правильно!\nТвое описание: {dw}\nПравильное описание: {dv}"
    else:
        eb=f"Неправильно!\nТвое описание: {dw}\nПравильное описание: {dv}"
    await dp.message.reply_text(eb)
    del g[dr]["dn"]

if __name__=="__main__":
    ec=Application.builder().token(a).build()
    ec.add_handler(CommandHandler("start",ab))
    ec.add_handler(CallbackQueryHandler(bq))
    ec.add_handler(MessageHandler(filters.PHOTO,cr))
    ec.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND,do))
    print("Bot started!")
    ec.run_polling()
```
